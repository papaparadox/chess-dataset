## Analysis Questions and Hypotheses (Step 2)

### Q1 - Does the higher-rated player win more often?
**Type:** Correlation between variables  
**Hypothesis:** Players with a higher rating than their opponent will more games. The winrate of the higher-rated player will be way above 50%

### Q2 - Does playing as white give a bigger advantage?
**Type:** Comparison between groups  
**Hypothesis:** White wins more games than black, since they have an advantage of playing the first move. White's win rate will be above 50% across all games

### Q3 - Do higher-rated games last longer?
**Type:** Correlation between variables  
**Hypothesis:** Games between higher-rated players will have more moves on average, as stronger players are less likely to blunder early and tend to play more positionally, leading to the endgame

### Q4 - Does the choice of opening affects the outcome?
**Type:** Impact of one variable on another  
**Hypothesis:** Certain openings will have a statistically higher win rate for white or black. Aggressive openings (such as Sicilian Defense or King's Gambit ) will result in more wins by checkmate or resignation of the opponent compared to positional ones.

### Q5 - Are rated games played differently from unrated games?
**Type:** Comparison between groups  
**Hypothesis:** Rated games will last longer and end in resignation or checkmate more often, while unrated ones will have more losses due to timeouts, meaning that players tend to take rated games more seriously.

## Clean and Transform the Data (Step 4)

```python
df = df.drop_duplicates() # cleaning dataset from duplicates

# dataset had 429 duplicates
```

```python
# Removing potential outliers for future analysis

df = df[df['turns'] > 1] # deleting the games where the opponent quit/resigned 
 # before the game even started

# dataset had 17 games with just 1 turn
```

```python
# converting the data types of 'created_at' and 'last_move_at' to correct DateTime

df['created_at'] = pd.to_datetime(df['created_at'], unit='ms')
df['last_move_at'] = pd.to_datetime(df['last_move_at'], unit='ms')

df[['created_at', 'last_move_at']].head()

# Now these 2 columns have correct date format
```

```python
df['victory_status'] = df['victory_status'].str.lower().str.strip()
df['victory_status'] = df['victory_status'].str.lower().str.strip()

# making sure that categories are standardised - all lowercase and without any whitespaces
```

```python
# renaming the columns for more clarity during the analysis

df = df.rename(columns={
    'white_rating': 'rating_white',
    'black_rating': 'rating_black',
    'opening_ply': 'opening_moves',
    'opening_eco': 'opening_eco_code',
})
```

```python
# New Features

df['rating_diff'] = df['rating_white'] - df['rating_black'] # rating difference between opponents
df['avg_rating'] = ((df['rating_white'] + df['rating_black']) / 2).round(0).astype(int)
```

```python
# Game length

df['game_length'] = pd.cut(
    df['turns'],
    bins=[0, 20, 40, 60, 100, 999],
    labels=['Very Short', 'Short', 'Medium', 'Long', 'Very Long']
)
```

```python
# Binary target for future ML Model

df['white_won'] = (df['winner'] == 'white').astype(int)
```

```python
# Encoding categories varialbes for ML Model

from sklearn.preprocessing import LabelEncoder

le_victory_status = LabelEncoder()
le_winner = LabelEncoder()

df['victory_status_enc'] = le_victory_status.fit_transform(df['victory_status'])
df['winner_enc'] = le_winner.fit_transform(df['winner'])
```
