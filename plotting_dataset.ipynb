```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('datasets/chess_games_clean.csv')
```

## Hypotheses and Questions Charts (tested outcomes)

### Q1 and Q2 - Who wins and how do games end?

```python
# Winners count
winner_counts = df['winner'].value_counts()
plt.pie(
    winner_counts,
    labels=winner_counts.index,
    autopct='%1.1f%%',
    colors=['turquoise', 'lightblue', 'blue'],
    startangle=90,
)
plt.title('Who Wins? (Overall)')
```

```python
# Victory Status

order = df['victory_status'].value_counts().index
sns.countplot(data=df, x='victory_status', order=order, palette='pastel')
plt.title('How Games End', fontsize=13)
plt.xlabel('Victory Status')
plt.ylabel('Number of Games')
```

**Insights:** White wins slightly more often than black (4% difference) consistently.
Resignation is the most common ending of the game, followed by mate and timeout.
Partial Q2 hypothesis explained - white does win more, but only minimally (4% difference in winrate)

### Q1 - Does the higher-rated player win?

```python
# Histogram — white ratings
sns.histplot(df['rating_white'], kde=True, color='blue', label='White', bins=40)
plt.title('Rating distribution amongst white pieces players')
plt.xlabel('Rating White')
```

```python
# Histogram - black ratings

sns.histplot(df['rating_black'], kde=True, color='tomato', label='Black', bins=40, alpha=0.6)
plt.title('Rating distribution amongst black pieces players')
plt.ylabel('Rating Black')
```

**Insights:** Rating distributions for white and black are close to being identical and are in the middle of 1500-1600 ELO, confirming that the matchmaking is balanced.
Hypothesis Q1 explained, the higher-rated player wins significantly more than 50% of the time.
